# House Price Prediction using Linear Regression

This notebook demonstrates how to predict house prices based on land square footage using a Linear Regression model.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load the dataset
df = pd.read_csv('dataset.csv')
print("Dataset Shape:", df.shape)
df.head()
```

### Data Visualization
Let's visualize the relationship between Area and Price.

```python
plt.xlabel('Area')
plt.ylabel('Price')
plt.scatter(df.area, df.price, color='red', marker='+')
plt.title('Area vs Price')
plt.show()
```

### Data Preprocessing
Splitting into X (features) and y (target).

```python
X = df[['area']] # Feature must be 2D array
y = df['price']
```

### Model Training

```python
model = LinearRegression()
model.fit(X, y)

# Show model parameters
print(f"Coefficient (m): {model.coef_[0]}")
print(f"Intercept (b): {model.intercept_}")
```

### Visualizing the Best Fit Line
Linear Regression calculates the line that minimizes the overall error. It doesn't hit every point perfectly. Let's see the line.

```python
plt.xlabel('Area')
plt.ylabel('Price')
plt.scatter(df.area, df.price, color='red', marker='+', label='Actual Data')
plt.plot(df.area, model.predict(df[['area']]), color='blue', label='Best Fit Line')
plt.legend()
plt.title('Actual vs Predicted')
plt.show()
```

### Prediction Interface
Enter the square footage below to get a prediction.

```python
try:
    x = int(input("Enter the area of land in square feet: "))
    
    # Predict
    # Creating a DataFrame to suppress sklearn feature name warning
    input_data = pd.DataFrame([[x]], columns=['area'])
    predicted_price = model.predict(input_data)[0]
    
    print(f"\nPredicted Price for {x} sq.ft: {predicted_price:.2f}")

except ValueError:
    print("Please enter a valid integer number.")
```
